```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

### Defining Driveshaft Properties

```python
yield_stress = 228*10**6              # Pa        maximum shear stress (yield strength)
required_torque = 10                 # Nm        what maximum torque will the driveshaft need to hande?
density = 2810                        # kg/m^3    density of the material
length = 0.3                          # m         length of the driveshaft
```

### Defining Simulation Parameters

```python
max_outer_diameter = 0.4        # meters
num_data_points = 500
```

### Calculating values for the inner diameter (d) that meet the torque requirement


```python
# The minimum valid outer diameter occurs when the shaft is solid:
D_min = ((required_torque*16) / (yield_stress*np.pi))**(1/3)

# Test to make sure there are valid solutions for this minimum diameter
if D_min >= max_outer_diameter: 
    sys.exit('Warning: No Solutions for this range of diameters')

D = np.linspace(D_min, max_outer_diameter, num_data_points)  # make an array of possible outer diameters

d = ( D**4 - ((16*required_torque*D)/(np.pi*yield_stress)) )*0.25  # Then find each corresponding inner Diameter

plt.plot(D, d)
plt.xlabel('Outer Diameter')
plt.ylabel('Inner Diameter')
```

### Now find the mass as a function of the inner and outer diameter

```python
Mass = (np.pi*(D/2)**2 - np.pi*(d/2)**2)*length*density

plt.plot(D, Mass)
plt.xlabel('Outer Diameter')
plt.ylabel('Mass')
```

### Now find the rotational inertia as a function of the inner and outer diameter

```python
I = 0.5*Mass*((D/2)**2 + (d/2)**2)

plt.plot(D, I)
plt.xlabel('Outer Diameter')
plt.ylabel('Mass Moment of Inertia')
```

### In Conclusion
As you increase the outer diameter, both the mass and mass moment of inertia must increase in order to maintain the same
yield torque. In order to minimise both mass and mass moment of inertia, the smallest possible outer diameter must be used.
 This occurs when d = 0, so a solid driveshaft should be used.
